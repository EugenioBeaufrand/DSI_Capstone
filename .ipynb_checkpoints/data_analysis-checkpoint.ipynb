```python
df_user_hashtags['hashtags_set'] = 0
df_user_hashtags['hashtags_set'] = 0
for i in range(len(df_user_hashtags)):
    if type(df_user_hashtags['new'][i]) is list:
        df_user_hashtags['hashtags_set'][i] = list(set(df_user_hashtags['new'][i]))
df_user_hashtags
```

```python
df_user_hashtags['hash_cnt'] = 0
for i in range(len(df_user_hashtags)):
    df_user_hashtags['hash_cnt'][i] = len(df_user_hashtags['new'][i])
df_user_hashtags
```

```python
df_user_hashtags['hash_unique_cnt'] = 0
for i in range(len(df_user_hashtags)):
    df_user_hashtags['hash_unique_cnt'][i] = len(df_user_hashtags['hashtags_set'][i])
df_user_hashtags
```

```python
df_user_hashtags['hash_unique_normalized'] = 0
unique_max = max(df_user_hashtags['hash_unique_cnt'])
unique_min = min(df_user_hashtags['hash_unique_cnt'])
for i in range(len(df_user_hashtags)):
    df_user_hashtags.loc[i,'hash_unique_normalized'] = ((df_user_hashtags['hash_unique_cnt'][i] - unique_min)/(unique_max - unique_min)).astype(float)
df_user_hashtags
```

```python
df_user_hashtags['hash_normalized'] = 0
unique_max = max(df_user_hashtags['hash_cnt'])
unique_min = min(df_user_hashtags['hash_cnt'])
for i in range(len(df_user_hashtags)):
    df_user_hashtags.loc[i,'hash_normalized'] = ((df_user_hashtags['hash_cnt'][i] - unique_min)/(unique_max - unique_min)).astype(float)
df_user_hashtags
```

```python
df_user_hashtags['score'] = 0.7 * df_user_hashtags['hash_unique_normalized'] + 0.3 * df_user_hashtags['hash_normalized']
df_user_hashtags
```
